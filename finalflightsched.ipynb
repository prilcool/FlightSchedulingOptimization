```python
## Defining gate lock times
gatelockdallas = []
gatelockAustin = []
gatelockHouston = []

#### DEFINING TIME FUNCTIONS #####

# sample 0600
arrivalAndDept = [] ## recording arrival and dept times

allterminals = []
T1 = []
T2 = []
T3 = []
T4 = []
T5 = []
T6 = []
```

```python
def conv_time_into_minutes(time):
    hour = int(time[0:2])  #extracting hour from the military time format
    minutes = int(time[2:4])

    timesincemidnight = hour*totalminutes + minutes

    return timesincemidnight


totalminutes = 60
def timeInMilitary(min_dept_time):
    timeMilitary_Min = min_dept_time % totalminutes
    timeMilitary_hour = min_dept_time // totalminutes
    # converting values to String

    strMin = str(timeMilitary_Min)
    strHour = str(timeMilitary_hour)

    strHour1 = ''
    strMin1 = ''

    if len(strHour) == 1:
        strHour1 = '0' + strHour
    else:
        strHour1 = strHour

    #### same for minutes ###
    if len(strMin) == 1:
        strMin1 = '0' + strMin
    else:
        strMin1 = strMin
    ## concatenating both o/p hour and minute strings

    departure_time = strHour1 + strMin1

    return departure_time

def TimeOnGround(dest):
    if dest == 'HOU':
        Time_on_ground = 35
    elif dest == 'DAL':
        Time_on_ground = 30
    elif dest == 'AUS':
        Time_on_ground= 30
    return Time_on_ground
```

```python
### Defining gate availability

def check_availability(origin, dest, TimeOfArrival, TimeOfDeparture):
        global diff_between_arrivalANDdept
        if origin == 'AUS' and dest == 'DAL':
            for val in gatelockdallas:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                else:
                    diff_between_arrivalANDdept=0
        elif origin == 'AUS' and dest == 'HOU':
            for val in gatelockHouston:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                else:
                    diff_between_arrivalANDdept=0

        elif origin =='DAL' and dest == 'AUS':
            for val in gatelockAustin:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                else:
                    diff_between_arrivalANDdept=0
        elif origin == 'DAL' and dest == 'HOU':
            for val in gatelockHouston:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                else:
                    diff_between_arrivalANDdept=0

        elif origin == 'HOU' and dest == 'DAL':
            for val in gatelockdallas:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                else:
                    diff_between_arrivalANDdept=0
        elif origin == 'HOU' and dest == 'DAL':
            for val in gatelockdallas:
                arrival_time = val[0]
                dept_time = val[1]
                if TimeOfArrival>=arrival_time and dept_time>=TimeOfArrival:
                    diff_between_arrivalANDdept = dept_time - TimeOfArrival
                    diff_between_arrivalANDdept +=1
                    # ground time
                    break
                else:
                    diff_between_arrivalANDdept=0
               # print(diff_between_arrivalANDdept)
        return diff_between_arrivalANDdept



#### PRODUCING THE FLIGHT S
```

```python

def schedule_production(tail, flight_gate, origin, dest, firstflight_time,lastflight_time, total_flight_time):

    while firstflight_time < 1320:
        time_of_arrival = firstflight_time + total_flight_time

        if time_of_arrival > 1320:
            break
        else:
            if dest == 'DAL':
                Time_on_Ground = 30
            elif dest == 'AUS':
                Time_on_Ground = 30
            elif dest == 'HOU':
                Time_on_Ground = 35

            time_of_departure = firstflight_time + total_flight_time+ Time_on_Ground

            arrivalAndDept.append([time_of_arrival, time_of_departure])

            finaldept = timeInMilitary(firstflight_time)
            finalarr = timeInMilitary(time_of_arrival)


            T2.append([tail, origin, dest, finaldept, finalarr])

            firstflight_time = time_of_departure
            newDest = origin
            newOrigin = dest
            origin = newOrigin
            dest = newDest

    firstflight_time = 600


# dict_items([('AUS', [['G1', 'T1']]),
# ('DAL', [['G2', 'T2'], ['G3', 'T3']]),
#  ('HOU', [['G4', 'T4'], ['G5', 'T5'], ['G6', 'T6']])])
#
```
